```python
import polars as pl

item_df = pl.read_parquet(r"items.parquet")

result = item_df.with_columns([
    # step: 1-5 -> int
    pl.col("description")
    .str.extract(r"step\s*([1-5])", 1)
    .cast(pl.Int8)
    .alias("step"),

    # size -> map thành số
    pl.col("description")
    .str.extract(r"\b(Newborn|S|M|L|XL|XXL)\b", 1)
    .replace({
        "Newborn": 0,
        "S": 1,
        "M": 2,
        "L": 3,
        "XL": 4,
        "XXL": 5
    })
    .cast(pl.Int8)
    .alias("size"),

    # piece -> số
    pl.col("description")
    .str.extract(r"(\d+)\s*miếng", 1)
    .cast(pl.Int16)
    .alias("piece"),
]).select([
    "item_id",
    "category_l1",
    "category_l2",
    "category_l3",
    "step",
    "size",
    "piece"
])

print(result['size'].unique())
print(result['step'].unique())
print(result['piece'].unique())

result.write_parquet("output.parquet")
```

```python
import polars as pl

def compute_co_buy_column(target_item_id, df_trans):
    """
    Tính co_buy_score cho TẤT CẢ sản phẩm xuất hiện trong bảng giao dịch
    dựa trên một sản phẩm X (target_item_id) cho trước.
    """
    
    # 1. Lọc ra danh sách các 'phiên' (khách + ngày) mà sản phẩm X xuất hiện
    target_sessions = (
        df_trans.filter(pl.col("item_id") == target_item_id)
        .select(["customer_id", "updated_date"])
        .unique()
    )
    
    # 2. Tìm các sản phẩm A mua cùng trong các phiên đó và đếm số lần
    co_buy_counts = (
        df_trans.join(target_sessions, on=["customer_id", "updated_date"])
        .filter(pl.col("item_id") != target_item_id) # Loại bỏ chính X
        .group_by("item_id")
        .agg(pl.len().alias("co_buy_score"))
    )
    
    # 3. Tạo bảng kết quả cuối cùng từ danh sách item duy nhất trong transactions
    # và gắn điểm score vào.
    all_items = df_trans.select("item_id").unique()
    
    final_df = (
        all_items.join(co_buy_counts, on="item_id", how="left")
        .with_columns(
            pl.col("co_buy_score").fill_null(0) # Nếu không mua chung thì score = 0
        )
    )
    
    return final_df

df3=compute_co_buy_column("6904000000002", df2)
print(df3.filter(pl.col("co_buy_score") > 0))
```

```python
import polars as pl

def get_pair_co_buy_score(item_x, item_a, df_trans):
    # 1. Lấy danh sách các phiên mua hàng (duy nhất) của item_x
    sessions_x = (
        df_trans.filter(pl.col("item_id") == item_x)
        .select(["customer_id", "updated_date"])
        .unique()
    )
    
    # 2. Lấy danh sách các phiên mua hàng (duy nhất) của item_a
    sessions_a = (
        df_trans.filter(pl.col("item_id") == item_a)
        .select(["customer_id", "updated_date"])
        .unique()
    )
    
    # 3. Inner join để tìm các phiên mà người dùng mua CẢ 2 sản phẩm
    common_sessions = sessions_x.join(sessions_a, on=["customer_id", "updated_date"], how="inner")
    
    # 4. Trả về số lượng dòng (chính là co_buy_score)
    return common_sessions.height

item_1 = "6904000000002"
item_2 = "4783000000009" # Thay bằng ID sản phẩm thứ 2 bạn muốn kiểm tra

score = get_pair_co_buy_score(item_1, item_2, df2)

print(f"Điểm mua chung (co-buy score) của {item_1} và {item_2} là: {score}")
```
